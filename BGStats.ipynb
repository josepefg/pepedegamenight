# Task
Analyze the data in the file "BGStatsExport.json" to list all players who played this year, the number of games played by each, the number of different games played by each, the win percentage of each, and the total playing time of each player.

## Carregar os dados

### Subtask:
Carregar os dados dos arquivos JSON `plays` e `players` em DataFrames do pandas.


**Reasoning**:
Load the JSON file and create dataframes from the 'plays' and 'players' keys. Then display the head of each dataframe to verify the loading.



```python
import json
import pandas as pd

# Read the JSON file
with open('/content/BGStatsExport.json', 'r') as f:
    data = json.load(f)

# Convert the 'plays' data to a DataFrame
if 'plays' in data and isinstance(data['plays'], list):
    df_plays = pd.DataFrame(data['plays'])
    print("DataFrame df_plays created:")
    display(df_plays.head())
else:
    print("The JSON data does not contain a list under the key 'plays'.")
    df_plays = pd.DataFrame() # Create an empty DataFrame to avoid errors later

# Convert the 'players' data to a DataFrame
if 'players' in data and isinstance(data['players'], list):
    df_players = pd.DataFrame(data['players'])
    print("\nDataFrame df_players created:")
    display(df_players.head())
else:
    print("The JSON data does not contain a list under the key 'players'.")
    df_players = pd.DataFrame() # Create an empty DataFrame to avoid errors later
```

## Filtrar dados por ano

### Subtask:
Filtrar o DataFrame de jogos (`df_plays`) para incluir apenas as partidas jogadas no ano atual (2024).


**Reasoning**:
Convert the 'entryDate' column to datetime objects and filter the DataFrame for the year 2024.



```python
df_plays['entryDate'] = pd.to_datetime(df_plays['entryDate'])
df_plays_2024 = df_plays[df_plays['entryDate'].dt.year == 2025]
display(df_plays_2024.head())
print(f"Shape of df_plays_2024: {df_plays_2024.shape}")
```

## Extrair informações dos jogadores por partida

### Subtask:
Processar a coluna `playerScores` no DataFrame filtrado para extrair o `playerRefId` e a informação de `winner` para cada partida.


**Reasoning**:
Process the playerScores column to extract playerRefId and winner information for each play in the filtered 2024 DataFrame.



```python
play_player_data = []

for index, row in df_plays_2024.iterrows():
    play_id = index
    player_scores = row['playerScores']

    if isinstance(player_scores, list):
        for player_score in player_scores:
            player_ref_id = player_score.get('playerRefId')
            winner = player_score.get('winner')

            if player_ref_id is not None and winner is not None:
                play_player_data.append({
                    'play_index': play_id,
                    'playerRefId': player_ref_id,
                    'winner': winner
                })

df_player_plays = pd.DataFrame(play_player_data)
display(df_player_plays.head())
```

## Calcular estatísticas por jogador

### Subtask:
Calcular para cada jogador no ano atual: o número total de partidas jogadas, o número de jogos diferentes jogados, o número de vitórias e o tempo total de jogo.


**Reasoning**:
Calculate the total number of games played by each player, the number of wins, the total playing time, and the number of unique games played, then merge these statistics into a single DataFrame.



```python
# 1. Calculate total number of games played by each player
player_game_counts = df_player_plays.groupby('playerRefId').size().reset_index(name='total_games_played')

# 2. Calculate the number of wins for each player
player_win_counts = df_player_plays[df_player_plays['winner'] == True].groupby('playerRefId').size().reset_index(name='total_wins')

# 3. Calculate the total playing time for each player
# Merge df_player_plays with df_plays_2024 to get durationMin
player_play_duration = df_player_plays.merge(df_plays_2024[['durationMin']], left_on='play_index', right_index=True, how='left')
player_total_time = player_play_duration.groupby('playerRefId')['durationMin'].sum().reset_index(name='total_playing_time_min')

# 4. Calculate the number of unique games played by each player
# Merge df_player_plays with df_plays_2024 to get gameRefId
player_game_ref = df_player_plays.merge(df_plays_2024[['gameRefId']], left_on='play_index', right_index=True, how='left')
player_unique_games = player_game_ref.groupby('playerRefId')['gameRefId'].nunique().reset_index(name='total_unique_games')

# 5. Merge all the calculated statistics into a single DataFrame
player_stats = player_game_counts.merge(player_win_counts, on='playerRefId', how='left')
player_stats = player_stats.merge(player_total_time, on='playerRefId', how='left')
player_stats = player_stats.merge(player_unique_games, on='playerRefId', how='left')

# Fill NaN values in total_wins with 0 (for players with no wins)
player_stats['total_wins'] = player_stats['total_wins'].fillna(0).astype(int)

display(player_stats.head())
```

## Combinar dados de jogadores

### Subtask:
Juntar as estatísticas calculadas com os nomes dos jogadores do DataFrame `df_players`.


**Reasoning**:
Merge the player statistics with player names, select and reorder columns, calculate win percentage, and display the result.



```python
# Merge player_stats with df_players to get player names
df_player_stats_merged = player_stats.merge(df_players[['id', 'name']], left_on='playerRefId', right_on='id', how='left')

# Select and reorder columns
df_player_stats_final = df_player_stats_merged[['name', 'total_games_played', 'total_unique_games', 'total_wins', 'total_playing_time_min']]

# Calculate win percentage
df_player_stats_final['win_percentage'] = (df_player_stats_final['total_wins'] / df_player_stats_final['total_games_played']) * 100

# Handle potential division by zero (NaN or infinity)
df_player_stats_final['win_percentage'] = df_player_stats_final['win_percentage'].replace([float('inf'), float('-inf')], 0).fillna(0)

# Display the head of the resulting DataFrame
display(df_player_stats_final.head())
```

## Exibir os resultados

### Subtask:
Mostrar uma lista com as estatísticas de cada jogador para o ano atual.


**Reasoning**:
Display the final DataFrame containing the player statistics for the current year.



```python
# Sort the DataFrame by 'total_games_played' in descending order
df_player_stats_final_sorted = df_player_stats_final.sort_values(by='total_games_played', ascending=False)

display(df_player_stats_final_sorted)
```

```python
from IPython.display import display, HTML

def create_interactive_table(df):
    """
    Creates an interactive HTML table from a pandas DataFrame using DataTables.
    """
    # Convert the DataFrame to an HTML table string
    html_table = df.to_html(index=False, classes='display compact')

    # Create the full HTML structure with DataTables and custom CSS
    html_structure = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <title>Player Stats Table</title>
        <link rel="stylesheet" type="text/css" href="https://cdn.datatables.net/1.11.5/css/jquery.dataTables.css">
        <script type="text/javascript" charset="utf8" src="https://code.jquery.com/jquery-3.6.0.min.js"></script>
        <script type="text/javascript" charset="utf8" src="https://cdn.datatables.net/1.11.5/js/jquery.dataTables.js"></script>
        <style>
            body {{
                font-family: sans-serif;
            }}
            #player_stats_table {{
                border-collapse: collapse;
                width: 100%;
            }}
            #player_stats_table th, #player_stats_table td {{
                border: 1px solid #666; /* Darker border color for cells */
                padding: 8px;
                text-align: left;
            }}
            #player_stats_table th {{
                background-color: #4CAF50; /* Darker green background */
                color: white; /* White text color for better contrast */
            }}
            #player_stats_table tbody tr:nth-child(even) {{
                background-color: #f9f9f9;
                color: #333;
            }}
            #player_stats_table tbody tr:nth-child(odd) {{
                background-color: #f9f9f9;
                color: #333
            }}
            #player_stats_table tbody tr:hover {{
                background-color: #ddd;
            }}
        </style>
        <script>
            $(document).ready( function () {{
                $('#player_stats_table').DataTable();
            }} );
        </script>
    </head>
    <body>
        <h2>Player Statistics (2024)</h2>
        <table id="player_stats_table" class="display compact">
            <thead>
                <tr>
                    {''.join(f'<th>{col}</th>' for col in df.columns)}
                </tr>
            </thead>
            <tbody>
                {html_table.split('<tbody>')[1].split('</tbody>')[0]}
            </tbody>
        </table>
    </body>
    </html>
    """
    return html_structure

# Create the interactive HTML table
interactive_table_html = create_interactive_table(df_player_stats_final_sorted)

# To display the HTML directly in Colab (limited interactivity)
display(HTML(interactive_table_html))

# To save the HTML to a file (for sharing)
with open('player_stats_table.html', 'w') as f:
    f.write(interactive_table_html)

print("\\nHTML table generated and saved as 'player_stats_table.html'")
```
